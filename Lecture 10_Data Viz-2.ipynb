# Example 1: Pie Chart

Demo of a basic pie chart plus a few additional features.

In addition to the basic pie chart, this demo shows a few optional features:

    * slice labels
    * auto-labeling the percentage
    * offsetting a slice with "explode"
    * drop-shadow
    * custom start angle

Note about the custom start angle:

The default ``startangle`` is 0, which would start the "Frogs" slice on the
positive x-axis. This example sets ``startangle = 90`` such that everything is
rotated counter-clockwise by 90 degrees, and the frog slice starts on the
positive y-axis.

Please see: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html

```python
# The slices will be ordered and plotted counter-clockwise.
labellables = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [18, 37, 41, 13]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) # only "explode" the 2nd slice (i.e. 'Hogs')


fig1, ax1 = plt.subplots() #create a new figure object called fig1, and return a np array 
#containing the created subplot objects, where ax1 is the name. 
#You can also define the number of subplots by adjusting plt.subplots(), e.g., plt.subplots(2,2). 
ax1.pie(sizes, explode=explode, labels=labellables,colors=colors,autopct='%.1f%%',pctdistance = 0.6,
        shadow=True, startangle=0,frame = True)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
#fig1

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
#The following shows another way to implement the pie chart. Can you tell the difference?
#plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        #autopct='%.2f%%', pctdistance = 0.6, shadow=True, startangle=90, frame = True)
# Set aspect ratio to be equal so that pie is drawn as a circle.
#plt.axis('equal')
#plt.show()
```

```python
# The slices will be ordered and plotted counter-clockwise.
labellables = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [18, 37, 41, 13]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%.2f%%', pctdistance = 0.6, shadow=True, startangle=90, frame = True)
plt.axis('equal')
plt.show()
```

# Example 2: Line Chart

```python
x = np.linspace(-2, 2, 201)  #array of 201 points from -2 to 2
y = x**2
z = x**3
plt.plot(x, y, 'r--', label='Quadratic') 
# See http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot for different line styles and more
plt.plot(x, z, 'b-', label='Cubic')
plt.legend(loc = 'lower right')
# For more location options, see https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
plt.xlabel("X axis")
plt.ylabel("Y axis") 
plt.show()
```

### Simple demo with multiple subplots.

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(x1) * np.exp(-x1) #cosine values 
y2 = np.cos(x2)

plt.subplot(2, 1, 1) # subplot(nrows, ncols, plot_number)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2) # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()

```

# Example 3: Box Plot 

```python
#fake up some data
spread= np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) *100 -100
data =np.concatenate((spread, center, flier_high, flier_low),axis=0)
#The axis along which the arrays will be joined. If axis is None, arrays are flattened before use. Default is 0.
print spread
print center
print flier_high
print flier_low
print data
```

```python
# basic plot
#plt.boxplot(data,vert=False)

fig1, ax1 = plt.subplots()
ax1.set_title('Box Plot')
ax1.boxplot(data,vert=False) #function boxplot returns to a dictionary, contains the following keys. 
```

```python
#To show the outliser as green diamonds. 
#For more variation of boxplot, please see:https://matplotlib.org/api/_as_gen/matplotlib.pyplot.boxplot.html
green_diamond = dict(markerfacecolor='g', marker='D')
fig3, ax3 = plt.subplots()
ax3.set_title('Changed Outlier Symbols')
ax3.boxplot(data, flierprops=green_diamond)
```

# Example 4: Histogram

Please see: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html  or https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist

```python
# example data
mu = 100 # mean of distribution
sigma = 15 # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)

num_bins = 50

#plt.hist(x)
#plt.hist(x, num_bins)
plt.hist(x, num_bins, normed=1)
#plt.hist(x, num_bins, normed=True, facecolor='g')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ: $\mu=100$, $\sigma=15$')

plt.show()
```
