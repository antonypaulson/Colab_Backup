# Problem 2 NumPy (12 pts)
Create random vector of size 10. Each element is random generated from standard normal distribution. Please replace the maximum value by 0. Then print the vector. 

Solution: The problem has been solved in two ways

**Method 1:**

```python
#Creating the vector
vect = np.random.randn(10)
print 'The original vector is : '
print '\n',vect

#Finding and replacing the maximum element in the vector
m = np.max(vect)
np.place(vect, vect == m, 0)

print '\nThe vector with maximum value replaced by 0 is: '
print '\n',vect
```

**Method 2:**

```python
#Creating the vector
vect = np.random.randn(10)
print 'The original vector is : '
print '\n',vect

#Finding and replacing the maximum element in the vector
m = np.max(vect)
i = np.where(vect == m)
vect[i] = 0
print '\nThe vector with maximum value replaced by 0 is: \n'
print vect
```

# Problem 3 NumPy (12 pts)
Create two random uniform matrices of same shape (4,4). 
How to get the diagonal of a dot product of these two matrices? Notice that the diagonal is a 1D array containing all the diagonal elements of a square matrix. 

```python
#Creating the two uniform matrices
mat1 = np.random.rand(4,4)
mat2 = np.random.rand(4,4)

print '\n The first matrix is : \n', mat1
print '\n The second matrix is : \n', mat2


#Dot product of the two matrices
mat3 = np.dot(mat1, mat2)
print '\n Their dot product is : \n', mat3


#Finding the diagonal of the dot product
print '\n Their diagonal array is : \n', mat3.diagonal()
```

# Problem 4 Numpy (12 pts)
Random Walk

Consider a simple random walk strating from 0 with steps of 1 and -1 occuring with probablity 0.8 and 0.2 respectively. Then we implement a single random walk with 100 steps using the built-in random module in Numpy. Then plot out the cumulative steps using matplotlib.pyplot module. 

```python
#Importing the necesscary Libraries
import numpy as np
import matplotlib.pyplot as plt

#Creating an array of the steps taken based on probabilities
nsteps = 100
draws = np.random.choice(2,nsteps,p= [0.2,0.8])


#Replacing 0 with -1 in the array
steps = np.where(draws>0,1,-1)
print '\n The array containing the steps are :\n\n',steps

#Plotting the cumulative sum of the steps 
walk = steps.cumsum()

plt.title('Simple Random Walk ')
plt.xlabel('Position')
plt.ylabel('Steps')
plt.plot(walk)

```
