# Problem 1 Line Chart (25 pts)

In the file “AAPL-NFLX.txt”, there are minute-level stock prices of AAPL and NFLX, each occupying one line. For example, the frist line of prices are for AAPL; second line of prices are for NFLX.  

Please plot a line chart for AAPL and NFLX stock price in the same plot, with time as your x-axis and stock price being your y-axis. Make the line for AAPL red solid; make the line for NFLX blue solid. Please label each of the two stocks: AAPL or NFLX; and provide a legend for your figure. In this plot, the x-axis should be labeled as “Time in Minutes” and the y-axis should be labeled as “Stock Price”.

Hint: You can read the data in text using *pd.read_csv( )* method:
          
          pd.read_csv('AAPL-NFLX.txt',sep=',',header=None)
          
Your output should be like:

![title](Problem1.png)

```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Reading the text file into a dataframe and transposing the dataframe for simplicity
value_file = pd.read_csv('AAPL-NFLX.txt',sep=',',header=None)
frame = value_file.T

#Defining the list values 
Apple = frame[0]
Netflix = frame[1]

#Plotting as per requirements
plt.plot(Apple, 'r-', label='APPL')
plt.plot(Netflix, 'b-', label = 'NFLX')
plt.legend(loc = 'lower left')
plt.xlabel('Time in Minutes')
plt.ylabel('Stock Price')
plt.show()
```

# Problem 2 Line Charts in Multiple Plots (25 pts)

Please draw two plots in one figure, showing the stock price of AAPL and NFLX, respectively. Make the line for AAPL red solid with dot marker; make the line for NFLX blue solid with dot marker. Please label the two plots in this one figure as follows. Make a title of your figure as "Minute Stock Price Comparison".

```python
fig, axs = plt.subplots(2, 1)
fig.suptitle('Minute Stock Price Comparison', fontsize =16)
#Making a subplot for Apple prices
plt.subplot(2, 1, 1) 
plt.plot(Apple, 'r.-')
plt.xlabel('Time in Minutes')
plt.ylabel('Stock Price')

#Making a subplot for Netflix prices
plt.subplot(2, 1, 2) 
plt.plot(Netflix, 'b.-')
plt.xlabel('Time in Minutes')
plt.ylabel('Stock Price')

plt.subplots_adjust(hspace=0.5)
plt.show()
```

# Problem 3: Histogram (25 pts)

The file “Data_Weibull.txt” contains 10,000 data points, one on each line. Please plot a figure containing two subplots. Each plot is a normalized histogram based on the data in this file with number of bins = 50 and 100, respectively. Please label the x-axis as “Score” and y-axis as “Probability”. Display the figure title as “Test Score Distribution”. 

Hint: To read the data, consider using ``fhand=open('Data_Weibull.txt','r')``. 

Your output should look like the follows:
![title](Problem3.png)

```python
#Reading the data into a list
fhand=open('Data_Weibull.txt','r')
whole = fhand.read()
x = whole.split()
x = np.array(x)
#Converting all strings to integers and creating a new list
y = []
for i in x:
    b = int(i)
    y.append(b)
    
fig, axs = plt.subplots(1, 2)    
#Creating the first histogram with 50 bins
plt.subplot(1, 2, 1)
plt.hist(y, 50, normed=True, facecolor='r')
plt.xlabel('Score')
plt.ylabel('Probability')

#Creating the second histogram with 100 bins
plt.subplot(1, 2, 2)
plt.hist(y, 100, normed=True, facecolor='b')
plt.xlabel('Score')
plt.ylabel('Probability')

#Defining the title for the figure
fig.suptitle('Test Score Distribution', fontsize=16)

#Increase spacing between plots for clarity and displaying the plots
plt.subplots_adjust(wspace=0.5)
plt.show()
```

# Problem 4: Scatter Plot (25 pts)

Given "FacebookFriends.xlsx", please read the data of "Hours_per_week" and data of "Friends". Please draw a scatter plot where x-axis is "Hours_per_week" and y-axis is "Friends. Make the color and marker size of each data point is random. 

Hint: to read the data properly, consider using ``pd.read_excel()``. 


You output should look like the following:
![title](Problem4.png)

```python
import pandas as pd
#Reading data into a frame
frame = pd.read_excel('FacebookFriends.xlsx')
x = frame['Hours_per_week']
y = frame['Friends']
# Creating the scatterplot with N = 33 points
N = 33
#Randomising color and size of each datapoint
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha =0.8) 
```
